```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
df_AENA_pasajeros = pd.read_pickle("./raw_df_AENA_pasajeros.pkl")
df_AENA_operaciones = pd.read_pickle("./raw_df_AENA_operaciones.pkl")
df_AENA_mercancias = pd.read_pickle("./raw_df_AENA_mercancias.pkl")
```

```python
aeropuerto = "BIO"
```

```python
display(df_AENA_pasajeros[aeropuerto].loc[[2020,2021]])
display(df_AENA_pasajeros[aeropuerto].loc[[(2019,"Enero"),(2021,"Enero")]])
display(df_AENA_pasajeros[df_AENA_pasajeros.index.get_level_values("Mes") == 'Enero']["BIO"])
display(df_AENA_pasajeros[df_AENA_pasajeros.index.get_level_values("Mes").isin(['Enero', 'Febrero'])]["BIO"])
```

```python
ax = plt.gca()
df_AENA_pasajeros[aeropuerto].loc[[2016,2017,2018,2019,2020,2021]].plot(kind='line', color=(random.uniform(0, 1), random.uniform(0, 1), random.uniform(0, 1)), label=aeropuerto, ax=ax, figsize=(12,6))
plt.ylabel('Pasajeros')
ax.legend()
plt.show()
```
